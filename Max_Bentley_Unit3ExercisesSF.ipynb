## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**:

Describe what processes "generate" each of the previous discrete distributions models. Also report their support.

Here's a spicier way to phrase the question: describe these distributions' maximum entropy conditions in english.

Hint: Make sure to read any of the: preliZ documentation, relevant wikipedia articles, or wolfram alpha.

The Bernoulli distribution simply plots the probability ($p$) of any given event succeeding or failing. The x-value of 1 is the chance of success, while the x-value of 0 is the chance of failure.

The Binomial distribution plots the probability ($p$) of some number of the same event occurring (independently) given a certain number of chances ($n$) for an event to happen, plotting every integer value from 0 to n.

The Poisson distribution plots the probability of some number of the same event occurring independently given a fixed amount of time for the event to occur and the average (mean) number of times the event occurs ($μ$) in that amount of time, plotting every integer value from 0 to infinity.

The Geometric distribution plots the probability ($p$) of the first event occurring on any given chance for the event to occur, plotting every integer value from 1 to infinity.

The HyperGeometric distribution plots the probability of any number of the observation of any given event occurring, given the possible times the event can occur ($N$), the number of events observed occurring ($k$) and the number of observances ($n$). It plots every integer value from either 0 or $n-N+k$ (whichever is greater) to $k$ or $n$ (whichever is smaller)

The Categorical distribution just plots each value given in a table by their index value starting from 0 and going until all values are plotted.
